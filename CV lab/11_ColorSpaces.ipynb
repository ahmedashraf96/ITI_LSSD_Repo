```python
#Importing openCV library
import cv2 as cv
```

```python
#Variable used in looping operations
iterator = 0

#Variable used in getting specific values
value = 0

#Reading the image in BGR mode
bgr_image = cv.imread("../01Basics/Files/carkyo.jpg")

#Converting the image to HSV mode
hsv_image = cv.cvtColor(bgr_image, cv.COLOR_BGR2HSV)

#Seperating H, S and V components
h_component = hsv_image[:, :, 0]
s_component = hsv_image[:, :, 1]
v_component = hsv_image[:, :, 2]

#Looping over H component values and subtracting 5 from them
for i in h_component:

    #Subtracting 5 from H component
    value = h_component[iterator] - 5

    #If the value is less than zero set it to zero
    value[value < 0] = 0

    #Setting the H component value
    h_component[iterator] = value

    #Increasing the iterator by 1
    iterator += 1

#Resetting the iterator
iterator = 0

#Looping over S component values and Adding 10 to them
for i in s_component:

    #Adding 10 to S component
    value = s_component[iterator] + 10

    #If the value is more than 255 set it to 255
    value[value > 255] = 255

    #Setting the S component value
    s_component[iterator] = value

    #Increasing the iterator by 1
    iterator += 1

#Resetting the iterator
iterator = 0

#Looping over V component values and multiplying them by 1.25
for i in v_component:

    #Multiplying V component with 1.25
    value = v_component[iterator] * 1.25

     #If the value is more than 255 set it to 255   
    value[value > 255.0] = 255.0

     #Setting the V component value   
    v_component[iterator] = value

    #Increasing the iterator by 1
    iterator += 1

#Setting the new image
hsv_image[:, :, 0] = h_component
hsv_image[:, :, 1] = s_component
hsv_image[:, :, 2] = v_component

#Getting the new BGR image
new_bgr_image = cv.cvtColor(hsv_image, cv.COLOR_HSV2BGR)

#Showing the new bgr image
cv.imshow("Image", new_bgr_image)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
#Getting YCrCb image
YCrCb_image = cv.cvtColor(bgr_image, cv.COLOR_BGR2YCrCb)

#Getting the Y component and multiply it by 1.25
y_component = YCrCb_image[:, :, 0] * 1.25

#Checking if any of the Y component values is bigger than 255 to set it to 255
y_component[y_component > 255] = 255

#Setting the new Y component
YCrCb_image[:, :, 0] = y_component

#Getting the new BGR image
new_bgr_image = cv.cvtColor(YCrCb_image, cv.COLOR_YCrCb2BGR)

#Showing the new BGR image
cv.imshow("Image", new_bgr_image)
cv.waitKey(0)
cv.destroyAllWindows()
```
