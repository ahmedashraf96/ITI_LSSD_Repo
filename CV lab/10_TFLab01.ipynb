```python
#Importing tensorflow, numpy and random libraries
import tensorflow as tf
import numpy as np
import random
```

```python
#Creating x, y (inputs) and z(output) lists
x_list = []
y_list = []
z_list = []

#Creating input and output data
for i in range (10):

    #Getting random values from 1 to 100 in x and y
    x = int (random.randint(1, 100))
    y = int (random.randint(1, 100))

    #Getting the value of z based on x and y values
    z = int ( 5*x - 3*y + 7)

    #Saving the new data in x, y and z lists
    x_list.append(x)
    y_list.append(y)
    z_list.append(z)

#Converting from list to array datatype 
x_list= np.array(x_list,  dtype=float)
y_list= np.array(y_list,  dtype=float)
z_list= np.array(z_list,  dtype=float)

#Merging x and y arrays into 1-D array
merged_array = np.stack([x_list, y_list], axis=1)

#Creating model to train
model = tf.keras.Sequential(tf.keras.layers.Dense(units=1, input_shape=[2]))

#Compiling the model
model.compile(loss='mean_squared_error', optimizer=tf.keras.optimizers.Adam(0.1))

#Fitting the input and output data
model.fit(merged_array,z_list, epochs=500)

#Getting the weights
model.get_weights()
```
