# Corner Dtection

- Understand and Explore `Harris` corner detection.
- Using OpenCV, apply it to all images in Chess folder.

## Criteria

- Detect as much chessboard corners as you can in all images.

## Grade
- 12

```python
import numpy as np
import cv2
import time
from matplotlib import pyplot as plt

img = cv2.imread('chess.png')                     
cv2.namedWindow("hariss")
scale_percent = 20                 
width = int(img.shape[1] * scale_percent / 100)                
height = int(img.shape[0] * scale_percent / 100)               
dim = (width, height)                                          
                                               
resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA) 
                 
gray = cv2.cvtColor(resized,cv2.COLOR_BGR2GRAY)
gray = np.float32(gray)

time1 = time.time()
dst = cv2.cornerHarris(gray,2,3,0.04)
dst = cv2.dilate(dst,None)
time2 = time.time() - time1
# Threshold for an optimal value, it may vary depending on the image
resized[dst>0.01*dst.max()]=[0,0,255]

cv2.imshow("hariss",resized)
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.figure(figsize=(25,25), facecolor=(0, 1, 1))
plt.subplot(1, 2, 1), plt.imshow(img, 'gray')
plt.subplot(1, 2, 2), plt.imshow(resized, 'gray')

```
