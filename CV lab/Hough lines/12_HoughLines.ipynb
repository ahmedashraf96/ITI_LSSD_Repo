# Hough Lines Lab

## Required

Create you own playaround to explore houghlines

## Criteria

- Use any edge detection filter you need but `canny` is recommended.
- You must use `houghLinesP` funstion.
- Create a trackbar for each variable in the function.
- Find the suitable parametrs to apply on:
  - Building.jpg
  - Road.jpg
- Save you results as images with lines drawn.

## Grade

10

```python
import cv2 as cv
import numpy as np

def nothing(x):
    pass

building_imgC = cv.imread("Files/Hough/Building.jpg")
building_img = cv.cvtColor(building_imgC, cv.COLOR_BGR2GRAY)

road_imgC = cv.imread("Files/Hough/Road.jpg")
road_img = cv.cvtColor(road_imgC, cv.COLOR_BGR2GRAY)

buildingImg_canny = cv.Canny(building_img, 50, 200)
roadImg_canny = cv.Canny(road_img, 50, 200)

cv.namedWindow("Hough1")
cv.namedWindow("Hough2")

cv.createTrackbar("x", "Hough1", 0, 500, nothing)
cv.createTrackbar("y", "Hough1", 0, 150, nothing)
cv.createTrackbar("z", "Hough1", 0, 50, nothing)

cv.createTrackbar("x", "Hough2", 0, 500, nothing)
cv.createTrackbar("y", "Hough2", 0, 150, nothing)
cv.createTrackbar("z", "Hough2", 0, 50, nothing)

while True:
    image= building_imgC.copy()

    x = cv.getTrackbarPos("x", "Hough1")
    y = cv.getTrackbarPos("y", "Hough1")
    z = cv.getTrackbarPos("z", "Hough1")
    lines = cv.HoughLinesP(buildingImg_canny, 1, np.pi / 180, x, None, y, z)

    if lines is not None:
        for i in range(0, len(lines)):
            l = lines[i][0]
            cv.line(image, (l[0], l[1]), (l[2], l[3]), (0, 0, 255), 3, cv.LINE_AA)
            
    buildingImg_cannyC = cv.cvtColor(buildingImg_canny, cv.COLOR_GRAY2BGR)
    image = cv.addWeighted(image, 0.9, buildingImg_cannyC, 0.5, 0)

    cv.imshow("Hough1", image)
    k = cv.waitKey(1)
    if k == 27:
        cv.imwrite("Building_img.jpg", image)
        break

while True:
    image= road_imgC.copy()

    x = cv.getTrackbarPos("x", "Hough2")
    y = cv.getTrackbarPos("y", "Hough2")
    z = cv.getTrackbarPos("z", "Hough2")
    lines = cv.HoughLinesP(roadImg_canny, 1, np.pi / 180, x, None, y, z)

    if lines is not None:
        for i in range(0, len(lines)):
            l = lines[i][0]
            cv.line(image, (l[0], l[1]), (l[2], l[3]), (0, 0, 255), 3, cv.LINE_AA)
            
    roadImg_cannyC = cv.cvtColor(roadImg_canny, cv.COLOR_GRAY2BGR)
    image = cv.addWeighted(image, 0.9, roadImg_cannyC, 0.5, 0)

    cv.imshow("Hough2", image)
    k = cv.waitKey(1)
    if k == 27:
        cv.imwrite("Road_img.jpg", image)
        break

cv.destroyAllWindows()
```
